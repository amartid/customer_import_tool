```python
import csv

def csv_to_dict_list(file_path):
    with open(file_path, 'r') as csv_file:
        reader = csv.DictReader(csv_file)
        data = [row for row in reader]
    return data

# Example usage
file_path = 'customer_import.csv'
csv_data = csv_to_dict_list(file_path)
print(csv_data)
```
